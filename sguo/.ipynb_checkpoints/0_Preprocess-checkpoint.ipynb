```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/en.openfoodfacts.org.products.tsv", sep = "\t")
for col in df:
    if df[col].dtype == object:
#         print(col)
        df[col] = df[col].astype(str)
        df[col].replace("nan", "", inplace=True)
# display(df.head())
print(df.shape)
print(df.info(verbose=True))
display(df.head())
```

```python
split_elems = lambda x: pd.Series([single_elem for e in list(x) for single_elem in e.split(",")])

country_counts = split_elems(df["countries_en"]).value_counts()
print(country_counts.head(30))
# print(split_elems(df["countries_en"]).unique())
```

```python
print(split_elems(df["categories_en"]).value_counts() )
```

```python
# split_elems(df["main_category"]).value_counts().head(30)
main_category_counts = df["main_category"].value_counts()
main_category_counts.head(30)
```

**Potential labels to use for clustering**

*countries_en*, *main_categories*

Pick the most frequent classes of a label to use. 

```python
from sklearn.preprocessing import StandardScaler

nutr_vars = [col for col in df if col[-5:]=="_100g" \
                                         and col!="nutrition-score-fr_100g" \
                                         and col!="nutrition-score-uk_100g"]
label = "main_category" #for example. maybe we use a different one
top_vals = main_category_counts.index.tolist()[1:7]
print(top_vals)

df_scaled = df[[label]+nutr_vars].copy()
df_scaled[nutr_vars] = StandardScaler().fit_transform(df[nutr_vars])
df_scaled = df_scaled[df_scaled[label].isin(top_vals)]

display(df_scaled.head())
print(df_scaled.shape)
```

```python
#tentative
kept_nutr_vars = ["energy_100g","proteins_100g","sodium_100g",\
                  "fat_100g","carbohydrates_100g","sugars_100g","saturated-fat_100g"]
print(df_scaled[kept_nutr_vars].dropna().shape)
```
